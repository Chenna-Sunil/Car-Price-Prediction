```python
x = df.iloc[:, 1:]
x
```

```python
y = df.iloc[:, :1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.30, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
y_pred_df = pd.DataFrame(y_pred, columns= ['Predicted Values'])
y_pred_df
```

```python
y_test_df = pd.DataFrame(np.array(y_test), columns = ['Actual Values'])
y_test_df
    
```

```python
pd.concat([y_test_df, y_pred_df], axis =1)
```

```python
plt.title('Relation between the Actual Values and Predicted Values')
sns.regplot(x=y_test_df['Actual Values'], y=y_pred_df['Predicted Values'])
plt.show()
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test_df, y_pred_df)
score
```
